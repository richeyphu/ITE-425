```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://quotes.toscrape.com/"
```

```python
response = requests.get(url)
response.text
```

```python
soup = BeautifulSoup(response.text, 'lxml')
soup
```

```python
quotes = soup.find_all('span', class_='text')
quotes
```

```python
for quote in quotes:
    print(quote.text)
```

```python
authors = soup.find_all('small', class_='author')
authors
```

```python
for author in authors:
    print(author.text)
```

```python
tags = soup.find_all('div', class_='tags')
tags
```

```python
for tag in tags:
    print(tag.text)
```

```python
for i in range(0, len(tag)):
    tag = tags[i].find_all('a', class_='tag')
    for i in tag:
        print(i.text)
    print('\n----------\n')
```

```python
url = "https://scrapingclub.com/exercise/list_basic/"
```

```python
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'lxml')
soup
```

```python
items = soup.find_all('h4', class_='card-title')
items
```

```python
for item in items:
    print(item.text)
```

```python
for item in items:
    print(item.text.strip('\n'))
```

```python
prices = soup.find_all('h5')
prices
```

```python
for price in prices:
    print(price.text)
```

```python
pagination = soup.find_all('a', class_='page-link')
pagination
```
